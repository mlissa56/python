### Задание 1
Счётчик

Определите класс Counter, реализующий положительный десятичный счетчик, который может увеличивать или уменьшать свое значение в заданном диапазоне, включая границы диапазона. 

В классе должны быть предусмотрены следующие возможности:
- конструктор для инициализации счетчика значениями по умолчанию (стартовое значение и верхняя границы диапазона)
- метод для установки произвольного значения счётчика
- методы для увеличения и уменьшения текущего значения счетчика, по умолчанию на 1
- метод для получения текущего занчения счётчика
- метод для сброса счётчика

Все методы класса должны проверять выход текущего значения счетчика за допустимый диапазон. 

Создайте экземпляр счетчика со значениями по умолчанию и выведите на экран его начальные параметры. Далее проверьте его работу циклом в пределах диапазона, увеличивая его текущее значение от минимально возможного до максимального. Протестируйте работу всех методов

Добавьте возможность складывать 2 экземпляра класса Counter, таким образом чтобы результатом сложения также был объект типа Counter

```python
import random

class Counter:
    def __init__(self, min_value=0, max_value=80): # задание значений по умолчанию
        self.min_value = min_value
        self.max_value = max_value
        self.value = min_value

    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            raise ValueError("Выход за верхнюю границу")

    def decrement(self):
        if self.value > self.min_value:
            self.value -= 1
        else:
            raise ValueError("Выход за нижнюю границу")

    def get_value(self): # текущее значение
        return self.value

    def random_value(self): # произвольное значение
        self.value = random.randint(self.min_value, self.max_value)

    def reset(self): # сброс
        self.value = self.min_value
        
    def __add__(self, second):
        if isinstance(second, Counter): # проверяем на принадлежность классу Counter
            new_counter = Counter()
            new_counter.value = self.value + second.value
            if new_counter.value > new_counter.max_value:
                new_counter.value = new_counter.max_value
                print("Выход за верхнюю границу")
            if new_counter.value < new_counter.min_value:
                new_counter.value = new_counter.min_value
                print("Выход за нижнюю границу")
            return new_counter

counter = Counter()
print("Начальное значение счетчика по умолчанию:", counter.get_value())

counter = Counter(min_value=1, max_value=79)
print("Начальное заданное значение счетчика:", counter.get_value())

counter.increment()
print("Значение счетчика увеличенное на 1:", counter.get_value())

counter.decrement()
print("Значение счетчика уменьшенное на 1:", counter.get_value())

counter.random_value()
print("Значение счетчика устанавливаемое произвольно:", counter.get_value())

counter.reset()
print("Значение счетчика после сброса:", counter.get_value())

counter1 = Counter()
counter1.increment()
counter1.increment()
print("Значение counter1:", counter1.get_value())

counter2 = Counter()
counter2.increment()
print("Значение counter2:", counter2.get_value())

counter3 = counter1 + counter2
print("Значение counter3 после сложения:", counter3.get_value())

#counter = Counter(min_value=1, max_value=2)
#counter.increment()
#counter.increment()
#print("Верхняя граница:", counter.get_value()) # должно выводиться ValueError: Выход за верхнюю границу

counter = Counter(min_value=1, max_value=2)
counter.decrement()
counter.decrement()
print("Нижняя граница:", counter.get_value()) # должно выводиться ValueError: Выход за нижнюю границу
```
